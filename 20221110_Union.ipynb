```
is_first = True
for file in tbl_order_files:
    if is_first:
        order_all = pd.read_csv(file)   # 반복문 처음에는 file을 불러와서 DataFrame을 만든다
        is_first = False
    else:
        order_data = pd.read_csv(file)  # 반복문 두번째부터는 이미만들어진 DataFrame에 Concatenate한다
        order_all = pd.concat([order_all, order_data], ignore_index=True)

order_all
```

 - 가로로 결합하는 조인 : merge() 

```
# store_id를 기준으로 m_store의 데이터를 붙인다
order_data = pd.merge(order_data, m_store, on='store_id', how='left') # left인 order_data를 기준으로 붙인다
order_data
```

```
# m_store를 merge하면서 추가된 area_cd를 기준으로 m_area의 데이터를 붙인다
order_data = pd.merge(order_data, m_area, on='area_cd', how='left')
order_data
```

 - 테크닉 8 : 마스터가 존재하지 않는 코드에 이름을 설정
 

```
order_data[['takeout_flag', 'status']] # 이 두가지 컬럼의 데이터는 마스터가 제공되지 않으므로 어떤의미인지 모른다
```

```
# 두 컬럼의 데이터의 의미(마스터)를 부여하자
order_data['takeout_flag'].unique() # array([0, 1])
order_data['status'].unique() # array([9, 2, 1])
```

```
# 1. map(dict) 함수 적용
order_data['takeout_name'] = order_data['takeout_flag'].map({0:'delivery', 1:'takeout'})
order_data['status_name'] = order_data['status'].map({0:'주문 접수', 1:'지불 완료', 2:'배달 완료', 9:'주문 취소'})
order_data[['takeout_name', 'status_name']]
```

 - 테크닉 11 : 데이터 로딩하고 불필요 항목 삭제

```
import pandas as pd
order_data = pd.read_csv(output_file)
print(len(order_data))
order_data.head()
```

```
# 주문상태가 결제완료, 배달완료인 데이터만 남기자
order_data = order_data.loc[(order_data['status'] == 1) | (order_data['status'] == 2)] # .loc[] 안하고 바로 .[조건] 해도 됨
print(len(order_data))
order_data.columns
```

```
# 필요한 컬럼만 사용
analyze_data = order_data[['store_id','customer_id','coupon_cd',
                           'order_accept_date','delivered_date','total_amount',
                           'store_name','wide_area','narrow_area',
                           'takeout_name','status_name']]

print(analyze_data.shape)
analyze_data.head()
```
