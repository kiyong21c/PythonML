 - 테크닉 12 : 데이터 전체 이미지 파악

```
# 리스트에서 삭제된 항목 확인하는 함수

def compare_cols(columns1, columns2): # list 항목중 무엇이 삭제되었는지 비교해주는 함수
    df1 = pd.DataFrame({'before':columns1}) # 첫번째 리스트를 DF화
    df2 = pd.DataFrame({'after':columns2})  # 두번째 리스트를 DF화

    mask = df1['before'].isin(df2['after']) # df1의 항목이 df2에 있는지 확인 T/F

    df3 = df1[mask].rename(columns={'before':'after'}) # df1에 mask를 적용하여 df3로 지정
    df3 = df3.rename(columns={'before':'after'}) # df3 컬럼명 변경(inplace=True 사용가능)
    # df3.columns = ['after']                    # df3 컬럼명 지정(변경)

    return pd.merge(df1, df3, left_index=True, right_index=True, how='left').fillna('') # df1, df3를 합쳐서 보여주기
```

```
compare_cols(order_data.columns, analyze_data.columns)
```

 - 문자열(String)을 날짜형식으로 변환하는 두가지 방법
  > 1. Python 함수인 datetime.strptime(Str, '%Y-%m-%d') 사용

  > 2. Pandas 함수인 pd.to_datetime(Str) 사용

```
# "Pandas"의 to_datetime(str)함수 사용하여 datetime형식으로 변경
pd.to_datetime(analyze_data['order_accept_date']) # str → datetime 형식
```

```
# "Python"의 dt.strftime(datetime)함수 사용하여 다시 str형식으로 변경
pd.to_datetime(analyze_data['order_accept_date']).dt.strftime('%Y%m') # str → datetime → str
```

```
# 주문 연도+월 항목 추가
analyze_data['order_accept_date'] = pd.to_datetime(analyze_data['order_accept_date'])
analyze_data['order_accept_month'] = analyze_data['order_accept_date'].dt.strftime('%Y%m')

analyze_data[['order_accept_date', 'order_accept_month']].head()
```

```
# 배달 연도+월 항목 추가
analyze_data['delivered_date'] = pd.to_datetime(analyze_data['delivered_date'])
analyze_data['delivered_month'] = analyze_data['delivered_date'].dt.strftime('%Y%m')

analyze_data[['delivered_date', 'delivered_month']].head()
```

```
import matplotlib.pyplot as plt

# 월별 매출합계 시각화
month_data.sum().plot()
```

```
# 월별 평균매출 시각화
month_data.mean().plot()
```

```
# State_based plot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(month_data.sum())
plt.title('sum')

plt.subplot(1, 2, 2)
plt.plot(month_data.mean())
plt.title('mean')
```

```
# Object_oriented plot
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].plot(month_data.sum())
axes[1].plot(month_data.mean())

# State_based같이 plt.figure로 시작했지만 각각에 축을 변수로 지정해서 Object_oriented처럼 사용 가능 
plt.figure(figsize=(10, 5))
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)
ax1.plot(month_data.sum())
ax2.plot(month_data.mean())
```
