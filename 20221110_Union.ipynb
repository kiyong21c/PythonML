```
# 현재 디렉터리 검색
import glob
tbl_order_files = glob.glob(current_dir + '/tbl_order_*.csv') # 와일드카드 활용가능
tbl_order_files
```

```
is_first = True
for file in tbl_order_files:
    if is_first:
        order_all = pd.read_csv(file)   # 반복문 처음에는 file을 불러와서 DataFrame을 만든다
        is_first = False
    else:
        order_data = pd.read_csv(file)  # 반복문 두번째부터는 이미만들어진 DataFrame에 Concatenate한다
        order_all = pd.concat([order_all, order_data], ignore_index=True)

order_all
```

 - 가로로 결합하는 조인 : merge() 

```
# store_id를 기준으로 m_store의 데이터를 붙인다
order_data = pd.merge(order_data, m_store, on='store_id', how='left') # left인 order_data를 기준으로 붙인다
order_data
```

```
# m_store를 merge하면서 추가된 area_cd를 기준으로 m_area의 데이터를 붙인다
order_data = pd.merge(order_data, m_area, on='area_cd', how='left')
order_data
```
