 - 테크닉 12 : 데이터 전체 이미지 파악

```
# 리스트에서 삭제된 항목 확인하는 함수

def compare_cols(columns1, columns2): # list 항목중 무엇이 삭제되었는지 비교해주는 함수
    df1 = pd.DataFrame({'before':columns1}) # 첫번째 리스트를 DF화
    df2 = pd.DataFrame({'after':columns2})  # 두번째 리스트를 DF화

    mask = df1['before'].isin(df2['after']) # df1의 항목이 df2에 있는지 확인 T/F

    df3 = df1[mask].rename(columns={'before':'after'}) # df1에 mask를 적용하여 df3로 지정
    df3 = df3.rename(columns={'before':'after'}) # df3 컬럼명 변경(inplace=True 사용가능)
    # df3.columns = ['after']                    # df3 컬럼명 지정(변경)

    return pd.merge(df1, df3, left_index=True, right_index=True, how='left').fillna('') # df1, df3를 합쳐서 보여주기
```

  > 각 매장의 특징에 기반해 몇가지 그룹으로 나누어보자

```
# 매장id를 인덱스로 그룹화하여 각 연산값 중 total_amount에대한 연산값만 출력
analyze_data.groupby('store_id').agg(['size','mean','median','max','min'])['total_amount']
# size : 주문수
```

```
# groupby로 인해  인덱스로 설정된 store_id를 컬럼으로 : reset_index(drop=False)
store_clustering = analyze_data.groupby('store_id').agg(['size','mean','median','max','min'])['total_amount'].reset_index(drop=False)
store_clustering
```

```
store_clustering.drop(columns='store_id', inplace=True, axis=1)
```

```
import seaborn as sns
hexbin = sns.jointplot(x='mean', y='size', data=store_clustering, kind='hex')
```

```
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 스케일 조정(모델 훈련시 들어갈 데이터는 스케일 조정 필요)
sc = StandardScaler()
store_clustering_sc = sc.fit_transform(store_clustering)

# K-means 모델 객체생성
kmeans = KMeans(n_clusters=4, random_state=0)
# K-means 모델 훈련
clusters = kmeans.fit(store_clustering_sc) # 모델 훈련시 원본데이터X, 스케일조정된데이터O

# K-means 분류된 레이블 확인
store_clustering['cluster'] = clusters.labels_

store_clustering.head()
```

```
from ipywidgets import DatePicker
import datetime

order_data.loc[:, 'order_date'] = pd.to_datetime(order_data['order_accept_date']).dt.date # .dt.date : 날짜까지만 표현(날짜 단위로 조회할 예정이므로)

def order_by_date(val):
    clear_output()
    display(date_picker)
    pick_data = order_data.loc[(order_data['order_date'] == val['new']) &
                               order_data['status'].isin([1, 2])]
    display(pick_data.head())

date_picker = DatePicker(value=datetime.datetime(2021, 4, 1)) # 초기 표현 날짜
date_picker.observe(order_by_date, names='value')
display(date_picker)
```

```
# 초기 값 설정
min_date = datetime.date(2021, 4, 1)
max_date = datetime.date(2021, 4, 30) 

# display된 날짜선택이 완료되면 호출되는 함수
def order_between_date():       # set_min_date/set_max_date 함수호출시 호출되는 함수
    clear_output()
    display(date_picker_min)
    display(date_picker_max)
    pick_data = order_data.loc[(order_data['order_date'] >= min_date) &
                               (order_data['order_date'] <= max_date) &
                               (order_data['status'].isin([1, 2]))]
    display(pick_data.head())

# 가장 이른 날짜(기간 시작)를 설정하는 함수
def set_min_date(val):
    global min_date # "초기화된 전역변수"를 함수내에서 "변경하여" 사용할 때
    min_date = val['new']
    order_between_date()

# 가장 느린 날짜(기간 끝)를 설정하는 함수
def set_max_date(val):
    global max_date # 초기화된 전역변수를 함수내에서 사용할 때
    max_date = val['new']
    order_between_date()

date_picker_min = DatePicker(value=min_date)
date_picker_min.observe(set_min_date, names='value')
print("기간 시작")
display(date_picker_min)

date_picker_max = DatePicker(value=max_date)
date_picker_max.observe(set_max_date, names='value')
print("기간 종료")

display(date_picker_max)
```
